```python
import pandas as pd
```

```python
def merge_files(map_path, plate_path, media=None):
    map_df = pd.read_csv(map_path)
    plate_df = pd.read_csv(plate_path)
    mdf = pd.merge(map_df, plate_df, left_on = 'well', right_on = 'Well')
    mdf['media'] = media
    rename = {'SpotCountCh2': 'SpotCount_phrodo',
              'SpotTotalAreaCh2': 'SpotTotalArea_phrodo',
              'SpotAvgAreaCh2': 'SpotAvgArea_phrodo',
              'SpotTotalIntenCh2': 'SpotTotalInten_phrodo',
              'SpotAvgIntenCh2': 'SpotAvgInten_phrodo',
              'pHrodo_conc_ug': 'phrodo_conc_ug'
             }

    cols = ['Well', 'phrodo_conc_ug', 'DA_Tx', 'media', 'replicate',
        'XCentroid', 'YCentroid', 'ObjectAreaCh1']
    cols += sorted(rename.values())

    return mdf.rename(columns = rename)[cols]


```
